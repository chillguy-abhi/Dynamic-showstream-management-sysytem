```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
```

```python
try:
    df = pd.read_csv("/content/World Indicators.csv")
except FileNotFoundError:
    print("Error: 'World Indicators.csv' not found. Please ensure the file is correctly uploaded.")
    raise
```

```python
us_data = df[df['Country/Region'].str.contains('United States', na=False)].copy()
```

```python
X_us = us_data['Year']
y_us = us_data['Population Total']
X_us = sm.add_constant(X_us)
us_model = sm.OLS(y_us, X_us, missing='drop').fit()
```

```python
X_pred_us = pd.DataFrame({'const': [1], 'Year': [2015]})
predicted_population_2015 = us_model.predict(X_pred_us)[0]
```

```python
europe_data = df[df['Region'] == 'Europe'].copy()
```

```python
# Define dependent variable
y_europe = europe_data['Life Expectancy Female']

# Define the five independent (predictor) variables
predictor_cols = [
    'Birth Rate',
    'GDP',
    'Health Exp % GDP',          # Cleaned name for 'Health Exp % GDP'
    'Infant Mortality Rate',
    'Life Expectancy Male'
]
X_europe_full = europe_data[predictor_cols]

# Add Constant (Intercept)
X_europe = sm.add_constant(X_europe_full)
```

```python
europe_model = sm.OLS(y_europe, X_europe, missing='drop').fit()
```

```python
# Ensure the model is fitted (Part C code must run first to define europe_model)

# 1. Define Prediction Input (Fixes NameError)
new_data_input = {
    'const': [1],
    'Birth Rate': [0.03],               # 3% -> 0.03
    'GDP': [1000000000],                # 1 billion
    'Health Exp % GDP': [0.04],           # 4% -> 0.04
    'Infant Mortality Rate': [0.05],    # 5% -> 0.05
    'Life Expectancy Male': [80]
}
X_pred_europe = pd.DataFrame(new_data_input)

# 2. Predict and Round
X_pred_europe = X_pred_europe[europe_model.params.index] # Ensure column order matches model
predicted_le_female = europe_model.predict(X_pred_europe)[0]
predicted_le_female_rounded = round(predicted_le_female, 2)

# print(f"Predicted Life Expectancy Female (Rounded to 2 decimals): {predicted_le_female_rounded}")
print(f"Predicted US Population in 2015: {predicted_population_2015:,.0f}")
print(f"3. Predicted Life Expectancy Female (Rounded to 2 decimals): {predicted_le_female_rounded}")
```
